```python
# Lowering all the columns so there is no typo while mentioning column names
df.columns = [col.lower() for col in df.columns]
```

```python
# Check missing Value
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df = df.dropna(subset=["time_orderd"])
df.isnull().sum()
```

```python
df["delivery_person_age"] = df["delivery_person_age"].fillna(
    df["delivery_person_age"].median()
)

df["delivery_person_ratings"] = df["delivery_person_ratings"].fillna(
    df["delivery_person_ratings"].median()
)
```

```python
df.isnull().sum()
```

```python
df["multiple_deliveries"] = df["multiple_deliveries"].fillna(0)
```

```python
df["festival"] = df["festival"].fillna("No")
```

```python
df["city"] = df["city"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], dayfirst=True)
```

```python
df["time_orderd"] = pd.to_datetime(df["time_orderd"], format="%H:%M:%S", errors="coerce").dt.time
df["time_order_picked"] = pd.to_datetime(df["time_order_picked"], format="%H:%M:%S", errors="coerce").dt.time
```

```python
df["order_hour"] = pd.to_datetime(df["time_orderd"].astype(str)).dt.hour
```

```python
df[["time_orderd", "order_hour"]].head()
```

```python
df["time_slot"] = np.where(
    ((df["order_hour"] >= 11) & (df["order_hour"] <= 14)) |             ####Create Peak / Off-Peak Flag
    ((df["order_hour"] >= 18) & (df["order_hour"] <= 22)),
    "Peak",
    "Off-Peak"
)
```

```python
df[["order_hour", "time_slot"]].head(10)
```

```python
df = df.rename(columns={
    "weather_conditions": "weather",
    "road_traffic_density": "traffic_density",
    "type_of_vehicle": "vehicle_type",
    "type_of_order": "order_type",
    "time_orderd": "order_time",
    "time_order_picked": "order_picked_time",
    "time_taken (min)": "delivery_time_min"
})
```

```python
text_cols = [
    "weather",
    "traffic_density",
    "order_type",
    "vehicle_type",
    "festival",
    "city"
]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
df = df.dropna(subset=["order_time", "order_picked_time", "order_hour"])
```
